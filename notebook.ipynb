### Simple scatterplot with tooltip

By [Ben Welsh](http://palewi.re/who-is-ben-welsh/)

An example of how to add a tooltip to a scatterplot in Python's [Altair](https://altair-viz.github.io) charting library.

```python
import altair as alt
from vega_datasets import data
```

```python
cars = data.cars()
```

```python
alt.Chart(cars).mark_point().encode(
    x="Miles_per_Gallon",
    y="Horsepower",
    tooltip="Name"  # <-- That's all folks
)
```

Tim Broderick [asks a follow-up question](https://twitter.com/TimBroderick/status/998660297656295424). How can you add more data to the tooltip?

```python
alt.Chart(cars).mark_point().encode(
    x="Miles_per_Gallon",
    y="Horsepower",
    tooltip=['Name', 'Origin', "Horsepower", "Miles_per_Gallon"]
)
```
