### Simple scatterplot with tooltip

By [Ben Welsh](http://palewi.re/who-is-ben-welsh/)

An example of how to add a tooltip to a scatterplot in Python's [Altair](https://altair-viz.github.io) charting library.

```python
import altair as alt
from vega_datasets import data
```

```python
cars = data.cars()
```

```python
alt.Chart(cars).mark_point().encode(
    x="Miles_per_Gallon:Q",
    y="Horsepower:Q",
    tooltip="Name:N"  # <-- That's all folks
)
```
